```python
import config
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
x = np.arange(0, 2_000_000 + 1, 10_000)
y = np.maximum(config.EPS_START - x * config.DECAY_STEP_LENGTH, config.EPS_END)

plt.figure(figsize=(10, 6))
plt.title('Decaying eps during training')
plt.plot(x, y)
plt.xlabel('step')
plt.ylabel('eps (exploration probability)')
plt.annotate(f'max_eps: {y[0]:.3f}', xy=(x[0], y[0]))
plt.annotate(f'min_eps: {y[-1]:.2f}', xy=(x[-1] - 3e5, y[-1] + 5e-2))
plt.show()
```
