```python
import numpy as np
import matplotlib.pyplot as plt
from params import *

%matplotlib inline
```

```python
steps = np.arange(1, 500_000 + 1)
eps = EPS_END + (EPS_START - EPS_END) * np.exp(-EPS_DECAY_RATE * steps)
```

```python
plt.figure()
plt.title('Decaying eps')
plt.plot(steps, eps)
plt.xlabel('step')
plt.ylabel('eps (exploration probability)')
plt.annotate(f'max_eps: {eps[0]:.3f}', xy=(steps[0], eps[0]))
plt.annotate(f'min_eps: {eps[-1]:.2f}', xy=(steps[-1] - 1e5, eps[-1] + 5e-2))
plt.show()
```
